```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_df = pd.read_csv('owid-covid-data.csv')
covid_df.head()
```

```python
covid_df.loc[(covid_df['location'] == 'Spain') & (covid_df['total_cases'] > 0), :]
```

```python
covid_df.info()
```

```python
covid_df['date'] = pd.to_datetime(covid_df['date'])
covid_df['date'] = covid_df['date'].dt.month
covid_df.head()
```

```python
covid_df.loc[(covid_df['location'] == 'Spain') & (covid_df['date'] >= 3), :]
```

```python
covid_df.dropna(subset=['total_cases', 'total_deaths', 'new_cases', 'total_cases_per_million',
                       'new_cases_per_million'], inplace = True)
```

```python
covid_df['death_rate'] = (covid_df['total_deaths'] / covid_df['total_cases']) *100
covid_df.head()
```

```python
spain_df = covid_df.loc[covid_df['location'] == 'Spain', :]
spain_df
```

```python
spain_df.info()
```

```python
grouped_spain = spain_df.groupby('date').mean()
grouped_spain2 = spain_df.groupby('date').sum()
grouped_spain3 = spain_df.groupby('date')[['total_cases', 'total_tests', 'total_cases_per_million']].max()
```

```python
grouped_spain
```

```python
grouped_spain2
```

```python
grouped_spain3
```

```python
grouped_spain3['total_cases'] = grouped_spain3['total_cases'] / 1000000
fig, ax = plt.subplots()
x = grouped_spain3.index
y = grouped_spain3['total_cases']
ax.plot(x, y)

ax.set(xlabel='Month)', ylabel='Cases',
       title='Total COVID Cases - Spain (millions)')
ax.grid()

fig.savefig("total_cases.png")
plt.show()
```

```python
x1 = grouped_spain.index
x2 = grouped_spain.index
y1 = grouped_spain['new_cases_smoothed']
y2 = grouped_spain['new_deaths_smoothed']

fig, ax = plt.subplots()
plt.plot(x1, y1, label = "Cases")
plt.plot(x2, y2, label = "Deaths")
plt.xlabel('Month')
plt.ylabel('Average')
plt.title('Daily Average COVID Cases and Deaths Spain per Month- Smoothed')
plt.legend()

fig.savefig("daily_average.png")
plt.show()
```

```python
fig, ax = plt.subplots()
x = grouped_spain.index
y = grouped_spain['death_rate']
ax.plot(x, y)

ax.set(xlabel='Month)', ylabel='%',
       title='Average Death Rate per Month - %')
ax.grid()

fig.savefig("death_rate.png")
plt.show()
```

```python
grouped_spain3['total_tests'] = grouped_spain3['total_tests'] / 1000000
fig, ax = plt.subplots()
x = grouped_spain3.index
y = grouped_spain3['total_tests']
ax.plot(x, y)

ax.set(xlabel='Month)', ylabel='Tests',
       title='Total COVID Tests - Spain(millions)')
ax.grid()

fig.savefig("total_tests.png")
plt.show()
```

```python
europe_df = covid_df.loc[covid_df['continent']== 'Europe']
europe_df.head()
```

```python
grouped_europe = europe_df.groupby('date').mean()
grouped_europe2 = europe_df.groupby('date').sum()
grouped_europe3 = europe_df.groupby('date')[['total_cases_per_million', 'total_tests_per_thousand']].max()
```

```python
grouped_europe = grouped_europe.loc[(grouped_europe.index >=3) & (grouped_europe.index <=10), :]
```

```python
grouped_europe2
```

```python
grouped_europe3
```

```python
x1 = grouped_spain.index
x2 = grouped_europe.index
y1 = grouped_spain['new_cases_smoothed_per_million']
y2 = grouped_europe['new_cases_smoothed_per_million']

fig, ax = plt.subplots()
plt.plot(x1, y1, label = "Spain")
plt.plot(x2, y2, label = "Europe")
plt.xlabel('Month')
plt.ylabel('Cases')
plt.title('Daily Average New COVID cases per million - Smoothed')
plt.legend()

fig.savefig("daily_average_comparison.png")
plt.show()
```

```python
x1 = grouped_spain.index
x2 = grouped_europe.index
y1 = grouped_spain['new_tests_smoothed_per_thousand']
y2 = grouped_europe['new_tests_smoothed_per_thousand']

fig, ax = plt.subplots()
plt.plot(x1, y1, label = "Spain")
plt.plot(x2, y2, label = "Europe")
plt.xlabel('Month')
plt.ylabel('Tests')
plt.title('Daily Average New COVID Tests per thousand - Monthly')
plt.legend()

fig.savefig("tests_comparison.png")
plt.show()
```

```python
x1 = grouped_spain.index
x2 = grouped_europe.index
y1 = grouped_spain['stringency_index']
y2 = grouped_europe['stringency_index']

fig, ax = plt.subplots()
plt.plot(x1, y1, label = "Spain")
plt.plot(x2, y2, label = "Europe")
plt.xlabel('Month')
plt.ylabel('Stringency Index')
plt.title('Stringency Index - Monthly')
plt.legend()

fig.savefig("stringency_index.png")
plt.show()
```
